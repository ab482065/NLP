```python
language_codes = {
    'english':'en',
    'arabic':'ar',
    'german':'de',
    'indonesian':'id',
    'italian':'it',
    'portuguese':'pt',
    'spanish':'es',
    'french':'fr',
    'turkish':'tr',
    'danish':'da',
    'hindi':'hi'
}
# Translates column 'col_name' in dataframe 'df' from 'src_lang' to 'dest_lang=en'
# use language codes for specifying language
def translate_df(df, col_name, src_lang, dest_lang='en'):
    df['translated_text'] = df[col_name].apply(ts.google)
    print(df)
    return df
```

```python
for path in glob.glob('../data/all-processed/*.csv'):
    try:
        path_in_str = str(path)
        print(path_in_str)
        if 'indonesian_alfina' not in path_in_str \
                and 'german_ross' not in path_in_str \
                and 'arabic' not in path_in_str \
                and 'french' not in path_in_str \
                and 'english' not in path_in_str:
            df = pd.read_csv(path_in_str)
            df = translate_df(df, 'text', language_codes[path_in_str.split('_')[1]])
            print(df)
            df.to_csv('T_' + path_in_str.split('\\')[1])  
    except Exception as e:
        print(e.with_traceback)
        pass
```
